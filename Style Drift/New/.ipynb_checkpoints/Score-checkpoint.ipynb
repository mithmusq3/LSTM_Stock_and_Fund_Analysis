```python
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter(action="ignore", category = FutureWarning)
warnings.simplefilter(action="ignore")

yr = pd.read_excel('portfolio consolidation work 2015.xlsx')
Gp = pd.read_excel('Style Drift Characteristics.xlsx')
pd.options.display.max_columns = 9999


def remove(string):
    return string.replace(" ", "")


yr['Company_Name'] = yr['Company_Name'].str.upper()
yr['Company_Name'] = yr['Company_Name'].str.replace(' ', '')
yr['Company_Name'] = yr['Company_Name'].str.replace("AMALGAMATED", '')
yr['Company_Name'] = yr['Company_Name'].str.replace("(", "")
yr['Company_Name'] = yr['Company_Name'].str.replace("-", "")
yr['Company_Name'] = yr['Company_Name'].str.replace(")", "")
yr['Company_Name'] = yr['Company_Name'].str.replace("&", "AND")
yr['Company_Name'] = yr['Company_Name'].str.replace("CORPN.", "CORPORATION")
yr['Company_Name'] = yr['Company_Name'].str.replace("INDUSTRIAL", "INDL.")
yr['Company_Name'] = yr['Company_Name'].str.replace("COMPANY", "CO.")
yr['Company_Name'] = yr['Company_Name'].str.replace("INDUSTRIES", "INDS.")
yr['Company_Name'] = yr['Company_Name'].str.replace(".", "")

Gp['Company_Name'] = Gp['Company_Name'].str.upper()
Gp['Company_Name'] = Gp['Company_Name'].str.replace("AMALGAMATED", "")
Gp['Company_Name'] = Gp['Company_Name'].str.replace(' ', '')
Gp['Company_Name'] = Gp['Company_Name'].str.replace("(", "")
Gp['Company_Name'] = Gp['Company_Name'].str.replace("-", "")
Gp['Company_Name'] = Gp['Company_Name'].str.replace(")", "")
Gp['Company_Name'] = Gp['Company_Name'].str.replace("&", "AND")
Gp['Company_Name'] = Gp['Company_Name'].str.replace("CORPN.", "CORPORATION")
Gp['Company_Name'] = Gp['Company_Name'].str.replace("INDUSTRIAL", "INDL.")
Gp['Company_Name'] = Gp['Company_Name'].str.replace("COMPANY", "CO.")
Gp['Company_Name'] = Gp['Company_Name'].str.replace("INDUSTRIES", "INDS.")
Gp['Company_Name'] = Gp['Company_Name'].str.replace(".", "")

Gp['exists'] = np.nan
Gp['holdings'] = np.nan
for each in list(Gp.loc[Gp.Company_Name.isin(yr['Company_Name'])].index):
    matched_index = yr.index[yr['Company_Name'] == Gp['Company_Name'].iloc[each]].tolist()
    if matched_index:
        Gp['exists'].iloc[each] = 'Exists'
        Gp['holdings'].iloc[each] = yr['Holding Percentage'].iloc[matched_index[0]]
    
Gp['exists'].fillna('non-existent', inplace=True)
companyname = Gp['Company_Name'].tolist()
ch1 = Gp['Market Capitalisation'].tolist()
ch2 = Gp['EPS'].tolist()
ch3 = Gp['P/E'].tolist()
ch4 = Gp['P/B'].tolist()
ch5 = Gp['holdings'].tolist()
value = 0
for i, companyname in enumerate(companyname, 0):
    if pd.isnull(ch4[i]) or pd.isnull(ch1[i]) or pd.isnull(ch2[i]) or pd.isnull(ch3[i]) or pd.isnull(ch5[i]):
        continue
    else:
        value = (ch4[i]*ch1[i]*ch2[i]*ch3[i]*ch5[i])
        print(companyname)
        print(value)
        print("\n")
```
